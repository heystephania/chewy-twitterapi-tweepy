```python
#pulling the tweepy library to use tweepy functions 
import tweepy 
import pandas as pd
```

```python
#super secret authentication codes 
consumer_key = "KEY"
consumer_secret = "KEY"
access_token = "KEY"
access_token_secret = "KEY"
```

```python
#setting authentencation to pull data 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth)
```

```python
#using the cursor function to pull recent tweet by specified twitter account. Items 1 limits the number of tweets pulled 
cursor = tweepy.Cursor(api.user_timeline,id='Chewy',tweet_mode = "extended").items(1)
```

```python
#defining relevant values in list 
number_of_tweets = 3250 
tweets = []
likes = []
time = []

for i in tweepy.Cursor(api.user_timeline,id='Chewy',tweet_mode = "extended").items(number_of_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)  
```

```python
emojis = []
for i in tweets: 
    em_split_emoji = emoji.get_emoji_regexp().findall(i) #split(i)
    em_split_whitespace = [substr.split() for substr in em_split_emoji]
    #em_split = functools.reduce(operator.concat, em_split_whitespace)
    emojis.append(em_split_emoji)
```

```python
#creating data frame to parse JSON
df = pd.DataFrame({'tweets':tweets,'Emojis':emojis,'likes':likes,'time':time})
```
